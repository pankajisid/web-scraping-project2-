```python
def get_pages(url):
    response =requests.get(url)
    if response.status_code != 200:
        raise Exception("Failed to load the page {}".format(url))
    doc =BeautifulSoup(response.text, 'html.parser')
    return doc
```

```python
def get_quotes_and_tags(doc):
    quotes =doc.find_all('span', class_ ='text')
    for item in quotes:
        quote.append(item.get_text(strip=True) if item else None)
    tags = doc.find_all('meta', class_= 'keywords')
    for tag in tags:
        tags1.append(tag['content'])
    return quote,tags
    
```
